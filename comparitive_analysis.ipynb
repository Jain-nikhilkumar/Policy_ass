```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and metrics
models = ['Linear Regression (BoW)', 'Linear Regression (TF-IDF)', 'SVM (BoW)', 'SVM (TF-IDF)', 
          'MNB (BoW)', 'MNB (TF-IDF)', 'LSTM', 'BART']

accuracy = [0.75, 0.75, 0.58, 0.51, 0.75, 0.75, 0.86, 0.48]
precision = [0.75, 0.75, 0.74, 0.75, 0.75, 0.75, 0.86, 0.47]
recall = [0.75, 0.75, 0.58, 0.51, 0.75, 0.75, 0.86, 0.48]
f1_score = [0.75, 0.75, 0.50, 0.36, 0.75, 0.75, 0.86, 0.47]

x = np.arange(len(models))

# Create the figure and axes with unique shapes
fig, ax = plt.subplots(figsize=(12, 8))

# Plot different markers for each metric
ax.plot(x, accuracy, marker='o', linestyle='-', color='b', label='Accuracy', markersize=10, linewidth=2)
ax.plot(x, precision, marker='s', linestyle='--', color='r', label='Precision', markersize=10, linewidth=2)
ax.plot(x, recall, marker='^', linestyle='-.', color='g', label='Recall', markersize=10, linewidth=2)
ax.plot(x, f1_score, marker='d', linestyle=':', color='purple', label='F1-Score', markersize=10, linewidth=2)

# Add title and labels
ax.set_title("Model Evaluation Comparison", fontsize=20, fontweight='bold', color='darkblue')
ax.set_xlabel("Models", fontsize=15)
ax.set_ylabel("Scores", fontsize=15)

# Add grid and unique style
ax.grid(True, which='both', linestyle='--', linewidth=0.6, color='gray')

# Set x-axis labels
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45, ha='right', fontsize=12)

# Add a legend
ax.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
